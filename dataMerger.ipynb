```python
import os

directory = os.walk('/Users/Maxwell/PycharmProjects/Github/ProjectXLL/options')
fileList = []

for root, dirs, files in directory:
    for file in files:
        if file[-3:] in ['csv', 'xls'] or file[-4:] == 'xlsx':
            fileList.append(os.path.join(root, file))
```

```python
import pandas as pd
import datetime


csv = pd.DataFrame()
excel = pd.DataFrame()
options = pd.DataFrame()

for file in fileList: 
    
    if file[-3:] == 'csv':
        
        csvTemp = pd.read_csv(file)
        mTime = os.path.getmtime(file)
        cDate = datetime.datetime.fromtimestamp(mTime).date()
        #csvTemp.loc[:, 'cDate'] = cDate
        
        csvTemp.loc[:, 'cDate'] = datetime.date(2016, 11, 15)
        
        csv = pd.concat([csv, csvTemp])
    
    else:
        
        try:
            
            excTemp = pd.read_excel(file, None)
            
            mTime = os.path.getmtime(file)
     
            cDate = datetime.datetime.fromtimestamp(mTime).date()

            path, fileName = os.path.split(file)
            ticker = fileName.split('.')[0]
            
            excTemp['CALL'].loc[:, 'Type'] = 'Call'
            excTemp['CALL'].loc[:, 'Ticker'] = ticker
            excTemp['CALL'].loc[:, 'cDate'] = cDate
            
            excTemp['PUT'].loc[:, 'Type'] = 'Put'
            excTemp['PUT'].loc[:, 'Ticker'] = ticker
            excTemp['PUT'].loc[:, 'cDate'] = cDate
            
            excTemp2 = pd.concat([excTemp['CALL'], excTemp['PUT']])
        
            excel = pd.concat([excel, excTemp2])
            
        except KeyError:
            continue
```

```python
csv2 = csv.loc[:,['cDate', 'ExpDate', 'StrikePrice', 'Ticker', 'Type', 'Last', 'IV']]

for date in set(csv.ExpDate):
        csv2.ix[csv2.ExpDate == date, 'ExpDate'] = datetime.datetime.strptime(date, '%b %d, %Y').date()

csv2.head()
```

```python
excel = excel.rename(columns={'Date':'ExpDate', 'ImpliedVolatility': 'IV', 'LastPrice':'Last', 'Strike':'StrikePrice'})

excel2 = excel.loc[:,['cDate', 'ExpDate', 'StrikePrice', 'Ticker', 'Type', 'Last', 'IV']]

for date in set(excel.ExpDate):
        excel2.ix[excel2.ExpDate == date, 'ExpDate'] = datetime.datetime.strptime(date, '%Y-%m-%d').date()

excel2.head()
```

```python
optionData = pd.concat([csv2, excel2])

optionData = optionData.reset_index(drop=True)
optionData.head()
```

```python
from pandas_datareader import data as web
from time import sleep
def getStockPrice(df):
    
    df2 = df.copy()
    
    for ticker in set(df.Ticker):   
        for date in set(df.loc[df.Ticker == ticker, 'cDate']):
            try:
                f = web.DataReader(ticker, 'yahoo', start=date, end=date)
                df2.loc[(df.Ticker == ticker) & (df.cDate == date), 'StockPrice'] = f.loc[date, 'Close']
            except:
                print(repr(ticker))
                break
    return df2

getStockPrice(optionData.head(20))
```

```python
optionData = getStockPrice(optionData)
```

```python
import numpy as np

optionData.loc[:, 'T'] = ((optionData.loc[:, 'ExpDate'] - optionData.loc [:, 'cDate']) / datetime.timedelta(days=1)).astype(float) / 365.25
optionData2 = optionData.loc[(optionData.loc[:, 'T'] > 0), :]
```

```python
from ImpliedVolatility import ImpliedVolatility as iv
sigma0 = 0.5
r = 0.005

df = optionData2
dfCopy = df.copy()
for i in df.index:
    dfCopy.ix[i, 'ImpliedVolatility'] = iv(df.ix[i, 'StockPrice'], df.ix[i, 'StrikePrice'], df.ix[i, 'T'], r, sigma0, df.ix[i, 'Last'], df.ix[i, 'Type']).bsmBisectionVol()

dfCopy.head()
```

```python
optionData2 = optionData2.dropna()
optionData2.to_csv('optionData.csv')
```

```python
tolerance = 0.01

dfCopy.loc[:, 'Difference'] = dfCopy.IV - dfCopy.ImpliedVolatility
europeanOptions = dfCopy.loc[np.fabs(dfCopy.Difference) < tolerance, :]
europeanOptions.to_csv("europeanOptions.csv")
europeanOptions.head()
```
