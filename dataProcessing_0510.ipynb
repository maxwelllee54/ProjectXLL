```python
import pandas as pd
import datetime
csv = pd.read_csv('CleanData.csv')
```

```python
optionData = csv.loc[:,['currentDate', 'ExpDate', 'StrikePrice', 'Ticker', 'Type', 'Last', 'IV']]

for d2 in set(optionData.currentDate):

    optionData.ix[optionData.currentDate == d2, 'currentDate'] = datetime.datetime.strptime(d2, '%m/%d/%Y').date()

for d in set(optionData.ExpDate):

    optionData.ix[optionData.ExpDate == d, 'ExpDate'] = datetime.datetime.strptime(d, '%m/%d/%Y').date()


optionData.head()
```

```python
from pandas_datareader import data as web
def getStockPrice(df):
    
    df2 = df.copy()
    
    for ticker in set(df.Ticker):   
        for date in set(df.loc[df.Ticker == ticker, 'currentDate']):
            try:
                f = web.DataReader(ticker, 'yahoo', start=date, end=date)
                df2.loc[(df.Ticker == ticker) & (df.currentDate == date), 'StockPrice'] = f.loc[date, 'Close']
            except:
                print(ticker)
                break
                

    return df2

getStockPrice(optionData.head(20))
```

```python
optionData = getStockPrice(optionData)
```

```python
optionData.loc[:, 'T'] = ((optionData.loc[:, 'ExpDate'] - optionData.loc [:, 'currentDate']) / datetime.timedelta(days=1)).astype(float) / 365.25
optionData2 = optionData.loc[(optionData.loc[:, 'T'] > 0), :]
optionData2 = optionData2.reset_index(drop=True)
```

```python
optionData2.to_csv('CleanData_StockPrice.csv')
```

```python
optionData2.head()
```

```python
from ImpliedVolatility import ImpliedVolatility as iv
sigma0 = 0.5
r = 0.005

df = optionData2.copy()
dfCopy = df.copy()
for i in df.index:
    impVol = iv(df.ix[i, 'StockPrice'], df.ix[i, 'StrikePrice'], df.ix[i, 'T'], r, sigma0,
                                   df.ix[i, 'Last'], df.ix[i, 'Type'])
    dfCopy.loc[i, 'ImpliedVolatility'] = impVol.bsmBrentq()[0]

dfCopy.head(10)
```

```python
import numpy as np
tolerance = 0.05

dfCopy.loc[:, 'Difference'] = dfCopy.IV - dfCopy.ImpliedVolatility
outputData = dfCopy.loc[np.fabs(dfCopy.Difference) < tolerance, :]
outputData.to_csv("europeanOptions_0510.05tol.csv")
outputData.head()
```
